```python
import numpy as np
from scipy.optimize import least_squares
from scipy.special import ellipeinc, ellipkinc

def calcular_a3a1_con_rango_ajustado(a2a1):
    def ecuacion(a3a1):
        a3a2 = a3a1 / a2a1
        sin2theta = ((1 - (a2a1) ** 2) / (1 - (a3a1) ** 2))

        # Verifica si sin2theta est fuera de los lmites de arcsin
        if np.any(sin2theta < 0) or np.any(sin2theta > 1):
            return np.inf  # Devolver un valor alto para no considerar esta solucin
        
        sintheta = np.sqrt(sin2theta)
        phi = np.arccos(a3a1)

        E = ellipeinc(phi, sin2theta)
        F = ellipkinc(phi, sin2theta)

        # Verifica si phi, m o sin(theta) estn fuera de rango
        if not np.isfinite(E).all() or not np.isfinite(F).all() or np.any(np.sin(phi) == 0):
            return np.inf  # Devolver un valor alto para no considerar esta solucin

        A1 = 2 * a2a1 * a3a1 * (F - E) / ((np.sin(phi)) ** 3 * sin2theta)
        A2 = 2 * a2a1 * a3a1 * (E - F * (1 - sin2theta) - a3a2 * sin2theta * np.sin(phi)) / ((np.sin(phi)) ** 3 * sin2theta * (1 - sin2theta))
        A3 = 2 * a2a1 * a3a1 * ((a3a2 ** (-1)) * np.sin(phi) - E) / ((np.sin(phi)) ** 3 * (1 - sin2theta))

        # Asegrate de que no haya divisiones por cero
        if np.any(np.isnan(A1)) or np.any(np.isnan(A2)) or np.any(np.isnan(A3)):
            return np.inf  # Devolver un valor alto para no considerar esta solucin

        lhs = (A1 - A2)
        rhs = (a3a1 ** 2) * A3 * (1 - (a2a1 ** (-2)))

        # Verifica que lhs y rhs sean valores finitos
        if not np.isfinite(lhs).all() or not np.isfinite(rhs).all():
            return np.inf  # Devolver un valor alto para no considerar esta solucin

        return lhs - rhs
    
    # Rango de valores iniciales para a3a1
    rango_inicial = [
        0.4944*(10**(-1)), 0.9682*(10**(-1)), 0.1416, 0.1835, 0.2228, 0.2594,
        0.2937, 0.3256, 0.3554, 0.3833, 0.4094, 0.4338, 0.4567, 0.4781,
        0.4983, 0.5172, 0.5351, 0.5519, 0.5677, 0.5828
    ]

    # Resolver para cada valor inicial y seleccionar el mejor resultado
    soluciones = []
    for a3a1_inicial in rango_inicial:
        try:
            solucion = least_squares(
                ecuacion, 
                a3a1_inicial, 
                bounds=(0, 1),
                xtol=1e-10,  # Tolerancia en el valor de la solucin
                gtol=1e-10,  # Tolerancia en el gradiente
                ftol=1e-10   # Tolerancia en el valor de la funcin
            ).x[0]
            soluciones.append(solucion)
        except ValueError as e:
            print(f"Error en least_squares con valor inicial {a3a1_inicial}: {e}")
    
    if not soluciones:
        return np.nan  # Si no hay soluciones vlidas, devuelve NaN
    
    # Seleccionar la solucin que mejor satisfaga la ecuacin
    errores = [np.abs(ecuacion(solucion)) for solucion in soluciones]
    mejor_solucion = soluciones[np.argmin(errores)]
    
    return mejor_solucion

# Valores de a2a1
a2a1 = np.linspace(0.05, 1, 40)

# Calcular a3a1 para cada valor de a2a1
a3a1_calculados_rango_ajustado = np.array([calcular_a3a1_con_rango_ajustado(a) for a in a2a1])

# Imprimir los resultados con ms decimales
np.set_printoptions(precision=10)
print(a3a1_calculados_rango_ajustado)
```

```python
TW=(A1 + A2*(a2a1**2) - 2*A3*(a3a1**2))/(2*(A1 + A2*(a2a1**2) + A3*(a3a1**2)))

#qn=kn*(1-(n/5))

#Omegamoo2=2*B12/qn

f=1/2 * (A1/(a2a1**(2/3) * a3a1**(2/3)) + A2/(a2a1**(-2/3) * a3a2**(2/3)) + A3/(a3a1**(-2/3) * a3a2**(-2/3)))
#_CocienteRoR3=((f*(1 - (2*TW)))**(3*n/(3-n)))

def _Omegamoo2(n,kn):
    qn=kn*(1-(n/5))
    _Omegamoo2=2*B12/qn
    return _Omegamoo2

def _CocienteRoR3(n,kn):   
    _CocienteRoR3=((f*(1 - (2*TW)))**(3*n/(3-n))) + 0*kn
    return _CocienteRoR3

def _Omegabarra2(n,kn):
    _Omegabarra2=_CocienteRoR3(n,kn) * _Omegamoo2(n,kn)
    return _Omegabarra2

def _Jmoo(n,kn):
    _Jmoo=(np.sqrt(3) / 10) * kn * (_Omegamoo2(n,kn))**(1/2) * ((a2a1**(-2/3)) * (a3a1**(-2/3)) + (a2a1**(2/3)) * (a3a2**(-2/3)))
    return _Jmoo

def _Jbarra(n,kn):
    #_CocienteRoR3=((f*(1 - (2*TW)))**(3*n/(3-n))) + 0*kn
    _Jbarra=_Jmoo(n,kn) * _CocienteRoR3(n,kn)**(-1/6)
    return _Jbarra

def _Omegagorro2(n,kn):
    qn=kn*(1-(n/5))
    _Omegagorro2=qn * (_CocienteRoR3(n,kn))**(-1) * _Omegabarra2(n,kn)
    return _Omegagorro2

def _Jgorro2(n,kn):
    qn=kn*(1-(n/5))
    _Jgorro2=qn/(kn**2) * (_CocienteRoR3(n,kn))**(1/3) * (_Jbarra(n,kn))**2
    return _Jgorro2

def _Ebarra(n,kn):
    _Ebarra=-(3-n)/(5-n) * _CocienteRoR3(n,kn)**(1/3) * f * (1 -((3-2*n)/(3-n))*TW)
    return _Ebarra
```

# ELIPSOIDES de RIEMANN-S INCOMPRESIBLES (n=0)

## Integrales elpticas

```python
import numpy as np
from scipy.optimize import least_squares
from scipy.special import ellipeinc, ellipkinc

# Definicin de parmetros
a2a1_vals = np.linspace(0.05, 1, 40)
Cbarra = 1.767

# Valores especficos de a3a1
a3a1_vals = np.array([
    0.4944*(10**(-1)), 0.9682*(10**(-1)), 0.1416, 0.1835, 0.2228, 0.2594,
    0.2937, 0.3256, 0.3554, 0.3833, 0.4094, 0.4338, 0.4567, 0.4781,
    0.4983, 0.5172, 0.5351, 0.5519, 0.5677, 0.5828
])

# Funciones A1, A2, A3
def calcular_A1_A2_A3(a2a1, a3a1):
    try:
        a3a2 = a3a1 / a2a1
        sin2theta = ((1 - (a2a1) ** 2) / (1 - (a3a1) ** 2))
        sintheta = np.sqrt(sin2theta)
        phi = np.arccos(a3a1)

        E = ellipeinc(phi, sin2theta)
        F = ellipkinc(phi, sin2theta)

        A1 = 2 * a2a1 * a3a1 * (F - E) / ((np.sin(phi)) ** 3 * sin2theta)
        A2 = 2 * a2a1 * a3a1 * (E - F * (1 - sin2theta) - a3a2 * sin2theta * np.sin(phi)) / ((np.sin(phi)) ** 3 * sin2theta * (1 - sin2theta))
        A3 = 2 * a2a1 * a3a1 * ((a3a2 ** (-1)) * np.sin(phi) - E) / ((np.sin(phi)) ** 3 * (1 - sin2theta))

        if not (np.isfinite(A1) and np.isfinite(A2) and np.isfinite(A3)):
            return np.array([np.nan, np.nan, np.nan])
        
        return np.array([A1, A2, A3])
    except Exception as e:
        return np.array([np.nan, np.nan, np.nan])

# Funcin para definir el sistema de ecuaciones
def sistema_ecuaciones(variables, a2a1, Cbarra):
    a3a1, Omegabarra, zetabarra = variables
    
    A1, A2, A3 = calcular_A1_A2_A3(a2a1, a3a1)
    
    if np.isnan(A1) or np.isnan(A2) or np.isnan(A3):
        return [1e10, 1e10, 1e10]
    
    B12 = A2 - ((A1 - A2) / (a2a1**2 - 1))
    
    if not np.isfinite(B12):
        return [1e10, 1e10, 1e10]
    
    eq1 = Omegabarra**2 + ((a2a1)**2)/(1 + (a2a1)**2) * zetabarra**2 - 2 * B12
    eq2 = (a2a1**2)/(1 + a2a1**2) * zetabarra * Omegabarra - (A1 - A2)/(1 - (a2a1**(-2))) + a3a1**2 * A3
    eq3 = Cbarra/np.pi - (a3a1**(-1/3)) * ((a3a1/a2a1)**(-1/3)) * (2 * Omegabarra + zetabarra)
    
    return [eq1, eq2, eq3]

# Resolver para cada valor de a2a1 usando least_squares
soluciones = []
for a2a1 in a2a1_vals:
    for a3a1 in a3a1_vals:
        rango_inicial = [
            [a3a1, 0.15, -0.05], [a3a1, 0.4, 0.1], [a3a1, 0.1, -0.2], [a3a1, 0.6, 0.0], [a3a1, 0.3, -0.1]
        ]
        
        for x0 in rango_inicial:
            solucion = least_squares(sistema_ecuaciones, x0, bounds=([0, 0, -0.4], [1, 0.8, 0.2]), args=(a2a1, Cbarra), xtol=1e-12, gtol=1e-12, ftol=1e-12)
            
            if solucion.success:
                soluciones.append(solucion.x)
                break

# Resultados
soluciones = np.array(soluciones)
np.set_printoptions(precision=10)
print(soluciones)
```
